```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
from sklearn.decomposition import PCA

bc_df = pd.read_csv("breast_cancer.csv")
cmc_df = pd.read_csv("cmc.csv")
iris_df = pd.read_csv("iris.csv")
```

```python
# Breast Cancer
bc_df_clean = bc_df.drop(columns=['id'])
label_encoder = LabelEncoder()
bc_df_clean['diagnosis'] = label_encoder.fit_transform(bc_df_clean['diagnosis'])

# Separar X e Y
X_bc = bc_df_clean.drop(columns=['diagnosis'])
Y_bc = bc_df_clean['diagnosis']

# Padronização
scaler = StandardScaler()
X_bc_scaled = scaler.fit_transform(X_bc)

# PCA
pca = PCA(n_components=2)
X_bc_pca = pca.fit_transform(X_bc_scaled)

# DataFrame final
bc_pca_df = pd.DataFrame(X_bc_pca, columns=['PC1', 'PC2'])
bc_pca_df['Diagnosis'] = Y_bc
bc_pca_df.head()
```

```python
# CMC
X_cmc = cmc_df.drop(columns=['ContraceptiveMethod'])
Y_cmc = cmc_df['ContraceptiveMethod']
X_cmc_encoded = pd.get_dummies(X_cmc, columns=['HusbandOccupation'])

# Normalização
minmax_scaler = MinMaxScaler()
X_cmc_scaled = minmax_scaler.fit_transform(X_cmc_encoded)

# DataFrame final
X_cmc_scaled_df = pd.DataFrame(X_cmc_scaled, columns=X_cmc_encoded.columns)
X_cmc_scaled_df['ContraceptiveMethod'] = Y_cmc
X_cmc_scaled_df.head()
```

```python
# Iris
X_iris = iris_df.drop(columns=['variety'])
Y_iris = iris_df['variety']

# Codificação da variável categórica
label_encoder = LabelEncoder()
Y_iris_encoded = label_encoder.fit_transform(Y_iris)

# Padronização
scaler = StandardScaler()
X_iris_scaled = scaler.fit_transform(X_iris)

# PCA
pca = PCA(n_components=2)
X_iris_pca = pca.fit_transform(X_iris_scaled)

# DataFrame final
iris_pca_df = pd.DataFrame(X_iris_pca, columns=['PC1', 'PC2'])
iris_pca_df['Variety'] = Y_iris
iris_pca_df.head()
```
